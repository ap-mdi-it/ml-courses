In dit labo maken we een verkennende analyse open data van Uber. Terloops maken we kennis met de [Pandas](https://pandas.pydata.org/) en [Plotly](https://plotly.com/graphing-libraries/) bibliotheken.

Bron: [Uber Data Analytics Dashboard](https://www.kaggle.com/datasets/gauravduttakiit/uber-pickups-in-ny-city)

```python
import os

import kagglehub
import pandas as pd
import plotly.express as px
```

```python
# Download latest version
path = kagglehub.dataset_download("gauravduttakiit/uber-pickups-in-ny-city")

# Load data into Pandas DataFrame
csv_file = os.path.join(path, "uber-pickups-in-new-york-city", "uber-raw-data-apr14.csv")
df = pd.read_csv(csv_file)
print("✅ Data loaded successfully!")
```

```python
# Convert "Date/Time" column to datetime type
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
df["Date/Time"].dtype
```

```python
# Plot a heatmap of the Uber pickups on a specific date
target_date = pd.to_datetime("2014-04-01").date()

fig = px.density_map(
    df.loc[df["Date/Time"].dt.date == target_date, :],  # April 1, 2014
    lat="Lat",
    lon="Lon",
    radius=10,
    map_style="open-street-map",
)
fig.show()
```
