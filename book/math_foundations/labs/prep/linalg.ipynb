# Lineaire Algebra

Maak de weergegeven objecten aan of pas de transformaties toe in NumPy. Print telkens het resultaat.

```python
import numpy as np
```

$$
\begin{align}
\pmb{a} &= \begin{bmatrix}
2 & 5 & 8 & 11
\end{bmatrix}^T \cr
\pmb{b} &= \begin{bmatrix}
1 & 3 & 5 & 7
\end{bmatrix}^T
\end{align}
$$

```python
a = np.array([2, 5, 8, 11])
b = np.array([1, 3, 5, 7])

print(a)
print(b)
```

$$
\pmb{B} = \begin{bmatrix}
6.0 & 7.0 \cr
4.0 & 3.0 \cr
5.0 & 2.0
\end{bmatrix}
$$

```python
B = np.array([[6.0, 7.0], [4.0, 3.0], [5.0, 2.0]])
print(B)
```

$$
\pmb{B}^T
$$

```python
print(B.T)
```

$$
B_{2,2}
$$

```python
print(B[1, 1])
```

$$
\pmb{b}+7
$$

```python
b + 7
```

$$
\begin{align}
\pmb{A} &= \begin{bmatrix}
1.1 & 2.5 \cr
3.2 & 4.4 \cr
5.1 & 6.9
\end{bmatrix} \cr
\pmb{A} &\odot \pmb{B}
\end{align}
$$

```python
A = np.array([[1.1, 2.5], [3.2, 4.4], [5.1, 6.9]])

print(A * B)
```

$$
\pmb{a}^T\pmb{b}
$$

```python
p = np.dot(a, b)  # or a @ b
print(p)
```

$$
\begin{align}
\pmb{c} &= \begin{bmatrix}
3.8 & 22.1
\end{bmatrix}^T \cr
&\pmb{B}\pmb{c}
\end{align}
$$

```python
c = np.array([3.8, 22.1])
p = np.dot(B, c)  # or B @ c
print(p)
```

$$
||\pmb{b}||_2
$$

```python
l2_norm = np.sqrt(np.dot(b, b))  # or np.linalg.norm(b)
print(l2_norm)
```

$$
C = diag(5, 7, 1, 4)
$$

```python
C = np.diag(np.array([5, 7, 1, 4]))
print(C)
```

$$
\pmb{D} = \pmb{I}_{7}
$$

```python
D = np.identity(7)
print(D)
```

$$
\begin{align}
\pmb{E} &= \begin{bmatrix}
1.1 & 2.5 & 3.0 \cr
3.2 & 4.4 & 3.5 \cr
5.1 & 6.9 & 1.2
\end{bmatrix} \cr
F &= \text{tril}(E)
\end{align}
$$

```python
E = np.array([[1.1, 2.5, 3.0], [3.2, 4.4, 3.5], [5.1, 6.9, 1.2]])
F = np.tril(E)
print(F)
```
